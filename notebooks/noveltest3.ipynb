```python
def plot_network(self):
    import networkx as nx
    import matplotlib.pyplot as plt

    G = nx.Graph()
    for d in self.result:
        if not d["character1_name"] or not d["character2_name"]:
            continue
        for n in range(2):
            pref = f"character{n+1}_"
            name = d[pref + "name"]
            if not G.has_node(name):
                nd = {k.replace(pref, ""): d[k] for k in d if k.startswith(pref)}
                G.add_node(name, **nd)
        G.add_edge(
            d["character1_name"],
            d["character2_name"],
            **{k: d[k] for k in d if not k.startswith("character")},
        )

    plt.figure(figsize=(8, 6))

    # Get edge widths based on relationship_likelihood attribute
    edge_widths = [len(G.edges[edge]["interactions"]) for edge in G.edges]

    # Define color mapping for 'relationship_happened'

    nx.draw(
        G,
        with_labels=True,
        # node_color=node_colors,
        node_size=500,
        # edge_color=edge_colors,
        width=edge_widths,
        font_size=16,
        font_color="black",
    )
    plt.title(f'Interactions in in {self.name.replace(".txt","")}')
    plt.savefig(
        os.path.join(
            PATH_DATA, f'interaction_graphs.{self.name.replace(".txt",".png")}'
        )
    )
    plt.show()
```

```python
plot_network(llm)
```
